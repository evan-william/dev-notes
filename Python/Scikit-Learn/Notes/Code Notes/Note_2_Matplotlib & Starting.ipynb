```python
from sklearn.datasets import make_blobs

"""
make_blobs is a data generator in scikit-learn that creates synthetic datasets for clustering or classification.
It’s especially useful when you want to test algorithms like K-Means, DBSCAN, or Gaussian Mixture Models without using a real dataset.

It generates:
- A set of points (X) scattered in clusters (blobs).
- The corresponding cluster labels (y) for each point.

Think of it as a tool to create fake but realistic cluster data.
"""
```

```python
# Usage Example
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate synthetic dataset
"""
make_blobs → generates synthetic data grouped into blobs (clusters).
n_samples=300 → creates 300 total data points.
centers=3 → generates 3 clusters (by default, the centers are chosen randomly, but you can also pass specific coordinates).
cluster_std=1.0 → standard deviation of each cluster → controls how “spread out” each cluster is (smaller = tighter clusters, larger = more scattered).
random_state=42 → ensures reproducibility (so every run gives the same result).
"""
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Plot the blobs
"""
plt.scatter → creates a 2D scatter plot.
X[:, 0] → the first column of X = x-axis (horizontal coordinate).
X[:, 1] → the second column of X = y-axis (vertical coordinate).
c=y → colors each point based on its cluster label (y). Points in the same cluster get the same color.
cmap="viridis" → colormap used for coloring. "viridis" gives a nice gradient of distinct colors.
s=50 → size of each point in the scatter plot (50 pixels).
"""
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="viridis", s=50)
plt.title("make_blobs Example: 3 Clusters")
plt.show()

# 3 Cluster center means from the total rows, pick 3 column randomly as center, and all the other rows that
# is close to the center will be pulled as a cluster
```

```python
# Easy Examples

# Generate 5000 sample points grouped into 5 clusters
# n_features=5 -> each point will have 5 features (a 5D dataset)
# If n_features is not given, default = 2 (a 2D dataset for easy plotting)
X, y = make_blobs(n_samples=5000, centers=5)
X
# X -> feature matrix (all the generated points and their coordinates in feature space)
# y -> cluster labels (which cluster each point belongs to: 0,1,2,3,4)
```

```python
# Prints the cluster labels
y
```

```python
# X[:, 0] -> use the first feature as x-axis
# X[:, 1] -> use the second feature as y-axis
# c=y -> color the points according to their cluster label (0,1,2,...)
plt.scatter(X[:, 0], X[:, 1], c=y)
```
